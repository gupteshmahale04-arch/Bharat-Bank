# Q1. Create a Student DataFrame

# Q8. Remove Duplicate Rows

## Problem Statement
Create the following DataFrame:

| Name | Age | City | Marks |
|------|----:|------|------:|
| Aman | 21 | Bhopal | 85 |
| Rahul | 22 | Indore | 72 |
| Aman | 21 | Bhopal | 85 |
| Neha | 20 | Delhi | 91 |

Remove duplicate rows using **drop_duplicates()**.

## Sample Input
```python
Use the table above.
```

## Sample Output
```python
Duplicate rows removed.
```

````python
# Write your code here
# Create the DataFrame with duplicates
dup_data = {
    "Name": ["Aman", "Rahul", "Aman", "Neha"],
    "Age": [21, 22, 21, 20],
    "City": ["Bhopal", "Indore", "Bhopal", "Delhi"],
    "Marks": [85, 72, 85, 91]
}
df_dup = pd.DataFrame(dup_data)

# Remove duplicate rows
df_cleaned = df_dup.drop_duplicates()
print(df_cleaned)
````

# Q9. Fill Missing Values

## Problem Statement
Create the following DataFrame:

| Name | Marks |
|------|------:|
| Aman | 85 |
| Rahul | NaN |
| Neha | 91 |
| Priya | NaN |

Fill missing values in the **Marks** column with **0** using **fillna()**.

## Sample Input
```python
Use the table above.
```

## Sample Output
```python
Missing values replaced with 0.
```

````python
# Write your code here
import numpy as np

# Create the DataFrame with missing values
missing_data = {
    "Name": ["Aman", "Rahul", "Neha", "Priya"],
    "Marks": [85, np.nan, 91, np.nan]
}
df_missing = pd.DataFrame(missing_data)

# Fill missing values with 0
df_missing["Marks"] = df_missing["Marks"].fillna(0)
print(df_missing)
````

# Q10. Final Practice

## Problem Statement
Create a DataFrame with the following data:

| Name | City | Marks |
|------|------|------:|
| Aman | Bhopal | 85 |
| Rahul | Indore | 72 |
| Neha | Delhi | 91 |
| Priya | Bhopal | 68 |
| Ravi | Indore | 95 |

Perform all of the following:
1. Display students with Marks > 80.
2. Sort the DataFrame by Marks.
3. Find the average Marks of each City.

## Sample Input
```python
Use the table above.
```

## Sample Output
```python
Required outputs displayed.
```

````python
# Write your code here
# Create the initial DataFrame
final_data = {
    "Name": ["Aman", "Rahul", "Neha", "Priya", "Ravi"],
    "City": ["Bhopal", "Indore", "Delhi", "Bhopal", "Indore"],
    "Marks": [85, 72, 91, 68, 95]
}
df_final = pd.DataFrame(final_data)

# 1. Display students with Marks > 80
print("--- Students with Marks > 80 ---")
print(df_final[df_final["Marks"] > 80])
print("\n")

# 2. Sort the DataFrame by Marks
print("--- Sorted by Marks ---")
print(df_final.sort_values(by="Marks"))
print("\n")

# 3. Find the average Marks of each City
print("--- Average Marks by City ---")
print(df_final.groupby("City")["Marks"].mean())
````
